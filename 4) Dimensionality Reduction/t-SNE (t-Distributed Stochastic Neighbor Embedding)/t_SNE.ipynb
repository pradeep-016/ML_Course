```python
import pandas as pd
import numpy as np
from scipy.io import mmread
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import plotly.express as px
```

```python
matrix = mmread('matrix.mtx').tocsc()
```

```python
genes = pd.read_csv('genes.tsv', sep='\t', header=None)
barcodes = pd.read_csv('barcodes.tsv', sep = '\t', header = None)
```

```python
normalize_matrix = np.log1p(matrix.toarray())
```

```python
scaler = StandardScaler()
scaled_matrix = scaler.fit_transform(normalize_matrix)
```

```python
tsne = TSNE(n_components = 3, random_state = 42,perplexity = 30, max_iter = 1000)

tsne_res = tsne.fit_transform(scaled_matrix)
```

```python
tsne_df = pd.DataFrame(tsne_res, columns = ['TSNE1', 'TSNE2', 'TSNE3'])
tsne_df['Cell'] = barcodes.iloc[:,0]
```

```python
fig = px.scatter_3d(
    tsne_df, x = 'TSNE1', y = 'TSNE2', z = 'TSNE3',
    color = 'Cell',
    title = 't-SNE Visualization of Glioblastoma',
    labels = {'TSNE1':"Dim 1", 'TSNE2': 'Dim 2', 'TSNE3': 'Dim 3'}
)
fig.show()
```

```python
n_clusters = 5
kmeans = KMeans(n_clusters = n_clusters, random_state = 42)
clusters = kmeans.fit_predict(tsne_res)
```

```python
tsne_k_df = pd.DataFrame(tsne_res, columns = ['TSNE1','TSNE2', 'TSNE3'])
tsne_k_df['Cluster'] = clusters
```

```python
fig = px.scatter_3d(
    tsne_k_df, x = 'TSNE1', y = 'TSNE2', z = 'TSNE3',
    color = 'Cluster',
    title = 't-SNE Visualization of Glioblastoma with Clusters',
    labels = {'TSNE1':"Dim 1", 'TSNE2': 'Dim 2', 'TSNE3': 'Dim 3', 'Cluster':'Cluster ID'}
)
fig.update_traces(marker = dict(size = 5))
fig.show()
```
