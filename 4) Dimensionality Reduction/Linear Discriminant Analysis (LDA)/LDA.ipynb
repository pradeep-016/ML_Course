```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import plotly.express as px
```

```python
data = pd.read_csv("Human_Recognition_with_smartphone.csv")
```

```python
data.head()
```

```python
X = data.drop(columns = ['subject', 'Activity'])
y = data['Activity']
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state = 42)
```

```python
lda = LDA(n_components = 3)
X_train_lda = lda.fit_transform(X_train,y_train)
X_test_lda = lda.transform(X_test)
```

```python
lda_df = pd.DataFrame(X_train_lda, columns = ['LDA1','LDA2','LDA3'])
lda_df['Labels'] = le.inverse_transform(y_train)
```

```python
fig = px.scatter_3d(
    lda_df,
    x = 'LDA1', y = 'LDA2', z = 'LDA3',
    color = 'Labels',
    title = 'Linear Discriminant Analysis',
    labels = {'Labels':'Activity'}
)
fig.show()
```

```python
rf = RandomForestClassifier(n_estimators = 100, random_state = 42)
```

```python
rf.fit(X_train_lda, y_train)
```

```python
y_pred = rf.predict(X_test_lda)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the Random Forest Classifier: {accuracy:.2f}")
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```
