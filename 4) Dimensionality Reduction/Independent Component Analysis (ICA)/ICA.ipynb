```python
import pandas as pd
from sklearn.decomposition import FastICA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import plotly.express as px
```

```python
data = pd.read_csv("PhishingData.csv")
data.head()
```

```python
data.columns = data.columns.str.strip()
```

```python
X = data.drop(columns=['index', 'Result'])
y = data['Result']
```

```python
binary_data = data[data['Result'].isin([-1, 1])]
```

```python
X_binary = binary_data.drop(columns=['index','Result'])
y_binary = binary_data['Result']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_binary)
```

```python
ica = FastICA(n_components = 3, random_state = 42)
X_ica = ica.fit_transform(X_scaled)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_ica, y_binary, test_size = 0.2, random_state = 42)
```

```python
lr = LogisticRegression(max_iter = 1000)
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the Random Forest Classifier: {accuracy:.2f}")
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

```python
ica_df = pd.DataFrame(X_ica, columns = ['ICA1', 'ICA2', 'ICA3'])
ica_df['Label'] = y_binary.values
```

```python
fig = px.scatter_3d(
    ica_df,
    x = 'ICA1',
    y = 'ICA2',
    z = 'ICA3',
    color=ica_df['Label'].map({-1: "Phishing", 1: "Legitimate"}),
    title = '3D Scatter Plot of Independent Component',
    labels = {'color':'Class'}
)
fig.show()
```
