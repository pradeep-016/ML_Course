```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
import plotly.express as px
```

```python
data = pd.read_csv('/content/telescope_data.csv')
```

```python
data.head()
```

```python
numerical_data = data.drop(columns=["Unnamed: 0", "class"])
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numerical_data)
```

```python
eps = 3
min_samples = 5
```

```python
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
clusters = dbscan.fit_predict(scaled_data)
```

```python
data['Cluster'] = clusters
```

```python
valid_clusters = clusters != -1
if len(set(clusters[valid_clusters])) > 1:
    silhouette = silhouette_score(scaled_data[valid_clusters], clusters[valid_clusters])
    print(f"Silhouette Score: {silhouette}")
else:
    silhouette = "Not applicable (less than 2 clusters)"
    print("Silhouette Score: Not applicable (less than 2 clusters)")
```

```python
print(data['Cluster'].unique())
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=numerical_data.columns)
```

```python
scaled_df['Cluster'] = clusters
```

```python
fig = px.scatter_3d(
    scaled_df,
    x=numerical_data.columns[0],
    y=numerical_data.columns[1],
    z=numerical_data.columns[2],
    color=scaled_df['Cluster'].astype(str),
    title="3D Visualization of DBSCAN Clusters",
    labels={'color': 'Cluster'}
)
fig.show()
```
